```python
import os
import pandas as pd
import dask.dataframe as da
import numpy as np
import json
import glob
```

```python
path = "data/2016/"
a = glob.glob(os.path.join(path,"*.csv"))
path = "data/2017"
b = glob.glob(os.path.join(path,"*.csv"))

req_files = a + b
```

```python
all_csv = [da.read_csv(f,error_bad_lines = False, dtype='unicode') for f in req_files]
df_merged = da.concat(all_csv).compute()
```

```python
to_consider=['FlightDate' ,'Quarter' ,'Year' ,'Month' ,'DayofMonth' , 'Origin' ,'Dest' ,'DepTime' ,'DepDel15' ,'CRSDepTime' ,'DepDelayMinutes' , 'OriginAirportID' ,'DestAirportID' ,'ArrTime' ,'CRSArrTime' ,'ArrDel15' ,'ArrDelayMinutes']
airports = ['EWR','IAH', 'ATL', 'CLT', 'DEN', 'PHX', 'LAX', 'JFK', 'DFW', 'SFO', 'SEA', 'ORD', 'MIA', 'MCO', 'LAS']
```

```python
flight_df = df_merged[to_consider]
final_df = flight_df
```

```python
final_df = final_df.dropna()
final_df = final_df[(final_df.Origin.map(lambda x: x in airports)) & (final_df.Dest.map(lambda x : x in airports))]
final_df = final_df.drop_duplicates()
final_df
```

```python
final_df.to_csv("data/flight_final.csv")
```
